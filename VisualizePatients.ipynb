# Medical Report Data Analysis and Visualization


```python
import json
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
import seaborn as sns
sns.set()  # This applies Seaborn's default style to your plots


# Set style
sns.set_palette('husl')

# Load the JSON data
with open('processed_patients.json', 'r') as f:
    data = json.load(f)

# Create DataFrame
df = pd.DataFrame(data)
df['birth_date'] = pd.to_datetime(df['birth_date'], format='%d.%m.%Y')
print('Data loaded successfully!')
df.head()
```

# Patient Data Analysis


```python
# Basic Statistics
print(f"Total number of patients: {len(df)}\n")

# Gender distribution
plt.figure(figsize=(10, 5))
gender_counts = df['gender'].value_counts()
sns.barplot(x=gender_counts.index, y=gender_counts.values)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

print("\nGender Distribution:")
print(gender_counts)
```

# ECOG Score Analysis


```python
# ECOG Score Distribution
plt.figure(figsize=(10, 5))
ecog_counts = df['ecog'].value_counts().sort_index()
sns.barplot(x=ecog_counts.index, y=ecog_counts.values)
plt.title('ECOG Score Distribution')
plt.xlabel('ECOG Score')
plt.ylabel('Count')
plt.show()

print("\nECOG Score Distribution:")
print(ecog_counts)
```

# Birth Year Analysis


```python
# Birth Year Distribution
plt.figure(figsize=(12, 6))
df['birth_year'] = df['birth_date'].dt.year
sns.histplot(data=df, x='birth_year', bins=20)
plt.title('Birth Year Distribution')
plt.xlabel('Birth Year')
plt.ylabel('Count')
plt.show()

print("\nBirth Year Statistics:")
print(df['birth_year'].describe())
```

# Tumor Status Analysis


```python
# Tumor Status Distribution
plt.figure(figsize=(12, 6))
tumor_counts = df['tumor_status'].value_counts()
sns.barplot(x=tumor_counts.index, y=tumor_counts.values)
plt.title('Tumor Status Distribution')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nTumor Status Distribution:")
print(tumor_counts)
```
